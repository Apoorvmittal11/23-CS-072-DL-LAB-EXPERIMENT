Fundamentals of tensor operations using Pytorch and Numpy.

```python
import numpy as np
import torch

print("NumPy version:", np.__version__)
print("PyTorch version:", torch.__version__)
```

Why tensors?

```python
import time
N = 1000000

list1 = list(range(N))
list2 = list(range(N))

tensor1 = torch.arange(N)
tensor2 = torch.arange(N)

start = time.time()
list_result = [list1[i] * list2[i] for i in range(N)]
end = time.time()

print("List multiplication time:", end - start, "seconds")

start = time.time()
tensor_result = tensor1 * tensor2
end = time.time()

print("Tensor multiplication time:", end - start, "seconds")
```

Create 1D,2D and 3D tensors using Pytorch and Numpy.

```python
#numpy
np_1d = np.array([1, 2, 3])

np_2d = np.array([[1, 2, 3],
                  [4, 5, 6]])

np_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print(np_1d)
print(np_2d)
print(np_3d)
```

```python
#pytorch
torch_1d = torch.tensor([1, 2, 3])

torch_2d = torch.tensor([[1, 2, 3],
                         [4, 5, 6]])

torch_3d = torch.tensor([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print(torch_1d)
print(torch_2d)
print(torch_3d)
```

Show basic operations : Element wise operations.

```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])
print(a + b)
print(a - b)
print(a * b)
print(a / b)
a_np = np.array([1, 2, 3])
b_np = np.array([4, 5, 6])
print(a_np + b_np)
print(a_np * b_np)
```

Indexing and slicing operations( Boolean Masking, extracting subtensor.. etc)

```python
print("Basic Indexing")
x = torch.tensor([[10, 20, 30],
                  [40, 50, 60]])

print(x[0])
print(x[:, 1])
print(x[1, 2])

print("\nExtracting Sub-tensors")
print(x[0:2, 1:3])

print("\nBoolean Masking")
y = torch.tensor([1, 5, 10, 15, 20])

mask = y > 10
print(mask)
print(y[mask])
y_np = np.array([1, 5, 10, 15, 20])
print(y_np[y_np > 10])
```

Use .view, .reshape .unsqueeze and squeeze function in pytorch

```python
#.view()
t = torch.arange(12)
print(t)

t2 = t.view(3, 4)
print(t2)

#.reshape()
t3 = t.reshape(4, 3)
print(t3)

#.unsqueeze()
a = torch.tensor([1, 2, 3])
print(a.shape)

b = a.unsqueeze(0)
print(b.shape)

c = a.unsqueeze(1)
print(c.shape)

#.squeeze()
d = torch.tensor([[1, 2, 3]])
print(d.shape)

print(d.squeeze().shape)
```

Compare with .reshape in numpy

```python
np_arr = np.arange(12)
np_reshaped = np_arr.reshape(3, 4)
print(np_reshaped)
```

Broadcasting

```python
print("pytorch")
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

y = torch.tensor([10, 20, 30])

print(x + y)

print("numpy")
x_np = np.array([[1, 2, 3],
                 [4, 5, 6]])

y_np = np.array([10, 20, 30])

print(x_np + y_np)
```

In place vs out of place operations

```python
#out of place operations
a = torch.tensor([1, 2, 3])
b = a + 5

print(a)
print(b)

#in of place operations
a = torch.tensor([1, 2, 3])
a.add_(5)

print(a)
```
